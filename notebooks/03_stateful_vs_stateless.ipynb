# StandardScaler: transformación individual y distribución

Este notebook muestra únicamente cómo usar `StandardScaler` en Scikit-Learn y cómo cambian los datos:

- a nivel de una observación individual,
- y a nivel de distribución para varias variables.

## 1. Cargar datos e importar librerías

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../data/raw/concrete_data.csv')

TARGET = 'concrete_compressive_strength'
X = df.drop(columns=TARGET)

print(f'Forma del dataset: {df.shape}')
X.head()
```

## 2. Ajustar `StandardScaler` con train y transformar los datos

`StandardScaler` aprende una media y desviación estándar por columna usando `fit(X_train)`. Luego usa esos valores para transformar cualquier dato con `transform(...)`.

```python
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = pd.DataFrame(
    scaler.transform(X_train),
    columns=X_train.columns,
    index=X_train.index,
 )

print('Primeras medias aprendidas por el scaler:')
print(pd.Series(scaler.mean_, index=X_train.columns).head())

print('\nPrimeras escalas aprendidas por el scaler:')
print(pd.Series(scaler.scale_, index=X_train.columns).head())
```

## 3. Visualización individual: una fila antes y después

Tomamos una muestra del conjunto de entrenamiento y comparamos sus valores originales con los valores escalados.

```python
sample_index = X_train.index[0]

comparacion_fila = pd.DataFrame({
    'original': X_train.loc[sample_index],
    'escalado': X_train_scaled.loc[sample_index]
})

comparacion_fila
```

```python
columnas_plot = ['cement', 'water', 'age']

fig, axes = plt.subplots(len(columnas_plot), 1, figsize=(10, 9), constrained_layout=True)

for ax, columna in zip(axes, columnas_plot):
    ax.bar(
        ['original', 'escalado'],
        [X_train.loc[sample_index, columna], X_train_scaled.loc[sample_index, columna]],
        color=['steelblue', 'tomato']
    )
    ax.set_title(f'Fila {sample_index} - {columna}')
    ax.set_ylabel('valor')

plt.show()
```

## 4. Visualización de distribución antes y después

Ahora comparamos la distribución de algunas variables antes y después del escalado. Después de aplicar `StandardScaler`, cada variable queda centrada alrededor de 0 y con dispersión cercana a 1.

```python
columnas_plot = ['cement', 'water', 'age']

fig, axes = plt.subplots(len(columnas_plot), 2, figsize=(12, 10), constrained_layout=True)

for i, columna in enumerate(columnas_plot):
    axes[i, 0].hist(X_train[columna], bins=20, color='steelblue', alpha=0.8)
    axes[i, 0].set_title(f'{columna} original')

    axes[i, 1].hist(X_train_scaled[columna], bins=20, color='tomato', alpha=0.8)
    axes[i, 1].set_title(f'{columna} escalada')

plt.show()
```

```python
nuevo_dato = pd.DataFrame({
    'cement': [540],
    'blast_furnace_slag': [0],
    'fly_ash': [0],
    'water': [162],
    'superplasticizer': [2.5],
    'coarse_aggregate': [1040],
    'fine_aggregate': [676],
    'age': [28]
})

nuevo_dato_escalado = pd.DataFrame(
    scaler.transform(nuevo_dato),
    columns=nuevo_dato.columns
)

print('Nuevo dato original:')
print(nuevo_dato)
print('\nNuevo dato escalado:')
print(nuevo_dato_escalado)
```

## 5. Resumen

- `fit(X_train)` aprende la media y desviación estándar de cada columna.
- `transform(X)` aplica esa referencia a cualquier conjunto de datos.
- En una fila individual, vemos cómo cada valor cambia de escala.
- En la distribución completa, vemos que las variables quedan centradas alrededor de 0.
